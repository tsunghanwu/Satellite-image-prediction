# Literature Study
## *i. Precipitation Nowcasting with Satellite Imagery*
### *Vadim Lebedev et al.*
* This paper focuses on forecasting short-range precipitation with satellite images (i.e. nowcasting)
* There are two parts in precipitation nowcasting, including wind estimation and precipitation field movement
* The data should have good spatial and temporal resolution, direct measurement of rainfall, and geological coverage
* UNet with binary crossentropy and dice loss as loss function is utilized 
* Data processing is required
> **Radar preprocessing**: Multiple radar observation could be aggregated on a single map and there are disagreements\
> **Projection**: Parallax shift between radar and satellite data\
> **Frame rate conversion**: Different periods of observation between satellite and radar\
> **Timeline adjustment**: Roshydromet and  EUMETSAT mark the time stamp and different timing (end/start of the scan)\
> **Additional features**: Add more features to the stellite images to provide an additional signal
* For performance comparison on predicting satellite images, there are 5 different models tested (MPE, PP, Pointwise, U-Net w/o GFS and w/GFS). U-Net with GFS shows the best accuracy and F1 score performance


## *ii. DeepRain: ConvLSTM Network for Precipitation Prediction Using Multichannel Radar Data*
### *Seongchan Kim et al.*
* This paper focuses on predicting precipitation with 3-dimensional and 4-channel weather radar data
* The 4-channel represent 4 different altitudes
* ConvLSTM is different version of LSTM, which embedded with convolution operation. It has been previously demonstrated on precipitation prediction with 3-dimensional and 1-channel data
* Stacked ConvLSTM cells are utilized, which could enhance model performance
* For performance comparison on predicting radar map, there are 4 different models tested (linear regression, FC-LSTM, ConvLSTM one-stacked and two-stacked). ConvLSTM with two-stacked shows the best RMSE performance (~ 23% drop)


## *iii. Convolutional LSTM Network: A Machine Learning Approach for Precipitation Nowcasting*
### *Xingjian Shi et al.*
* This paper focuses on precipitation nowcasting with spatiotemporal sequence of radar map 
* The precipitation nowcasting problem could be seem as a spatiotemporal sequence problem
* A encoding-forcasting structure with stacked ConvLSTM layers is utilized, which is build to address spatiotemporal sequence forecating problems
* ConvLSTM shows better cross entropy than FC-LSTM on Moving-MNIST dataset
* For performance comparison on prediction radar map sequence, there are 5 models tested (ConvLSTM, ROVER1/2/3, FC-LSTM). ConvLSTM shows the best MSE (and other metrics) performance

```python
def crop_image(image, x_min=300, x_max=400, y_min=100, y_max=200):

    """This function is to crop the image

    Args:
        image: input image
        x_min: the minimum of the x-coordinate (default = 300)
        x_max: the maximum of the x-coordinate (default = 400)
        y_min: the minimum of the y-coordinate (default = 100)
        y_max: the maximum of the y-coordinate (default = 200)

    Returns:
        Cropped image
    
    """
    img_array = img_to_array(image)
    img_array_crop = img_array[y_min:y_max, x_min:x_max, :]
    image_cropped = array_to_img(img_array_crop)
    
    return image_cropped
```

```python
def add_gray_avg(df):

    """This function is to add the averaged grayscale value to the summary dataframe

    Args:
        df: input dataframe

    Returns:
        Dataframe with an additional column of averaged grayscale value
  
    """

    gray_avg_list = []

    for filepath in df['Filepath']:
        image = load_img(filepath)
        image_cropped = crop_image(image)
        img_array = img_to_array(image_cropped)
        rgb_weights = [0.2989, 0.5870, 0.1140]
        gray_img_array = np.dot(img_array[:,:,:3], rgb_weights)
        avg = np.mean(gray_img_array)
        gray_avg_list.append(avg)
    
    df['Gray_avg'] = gray_avg_list

    return df

df = add_gray_avg(df)
```

# Data Processing
## Desciption
*   Create a function to extract the 3D grayscale array from file path
*   Create a function to generate array queue from summary dataframe
*   Create a function to generate original array sequence and shifted array squence  from array queue

```python
def path_to_array(filepath):

    """This function is to crop the images and transfomr to 3D grayscale array

    Args:
        filepath: File path of the image
  
    Returns:
        3-dimensional grayscale array

    """

    image = load_img(filepath)

    # Crop the image and convert to array
    image_cropped = crop_image(image)
    image_cropped_array = img_to_array(image_cropped)

    # Trasfrom to 3D grayscale array 
    rgb_weights = [0.2989, 0.5870, 0.1140]
    gray_img_array = np.dot(image_cropped_array[:,:,:3], rgb_weights)
    gray_img_array = gray_img_array.reshape(100, 100, 1)

    return gray_img_array
```

```python
def gen_array_queue(df, width=100, height=100):

    """This function is to pack the image arrays into image queue

    Args:
        df: Input dataframe
        width: width of the image
        height: height of the image

    Returns:
        Output image queue

    """

    n_images = len(df)
    width = 100
    height = 100

    img_queue = np.zeros((n_images, height, width, 1))

    for i, path in enumerate(df['Filepath']):
        img_array = path_to_array(path)
        img_queue[i, :, :, :] = img_array

    return img_queue
```

```python
def gen_img_seqs(img_queue, n_frames=10, sep=1, frame_shift=1, width=100, height=100):

    """This function is to generate image sequences

    Args:
        img_queue: input image queue
        n_frames: number of frames in a sequence
        sep: frame seperation between different sequences
        frame_shift: the number of frame shifts for the training target sequence
        width: width of images
        height: height of images

    Returns:
        img_seq_X: image sequences for training
        img_seq_y: image sequences for training target

    """
    
    n_samples = (len(img_queue) - n_frames - frame_shift +1) // sep
    img_seq_X = np.zeros((n_samples, n_frames, height, width, 1))
    img_seq_y = np.zeros((n_samples, n_frames, height, width, 1))

    for i in range(n_samples):
        j = i*sep
        img_seq_X[i, :, :, :, :] = img_queue[j:j+n_frames, :, :, :]
        img_seq_y[i, :, :, :, :] = img_queue[j+frame_shift:j+n_frames+frame_shift, :, :, :]

    return img_seq_X, img_seq_y
```

```python
def plot_seq(img_seq, sample, n_frames=10, width=100, height=100):

    """This function is to plot the image sequence

    Args:
        img_seq: input image sequence
        sample: the desired sample for ploting
        n_frames: number of frames per image
        width: width of the image
        height: height of the image
  
    """

    plt.figure(figsize=(14,7))
    for i in range(n_frames):
        plt.subplot(2,5,i+1)
        plt.imshow(img_seq[sample][i].reshape(height,width), cmap='gray')
        plt.text(5, 10, 'frame={}'.format(i+1), size=12, color='tab:red')
    plt.tight_layout() 
```

# Model Building
## Description
*   The model is built on multiple ConvLSTM layers
*   MSE as loss function and Adam as optimizer
*   30% of the training data are used for validation
*   Batch size is set to 2 due to limited GPU resource

```python
# Build ML model
model = Sequential()

model.add(ConvLSTM2D(filters=40, kernel_size=(3, 3),input_shape=(None, 100, 100, 1), padding='same', return_sequences=True))
model.add(BatchNormalization())

model.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), padding='same', return_sequences=True))
model.add(BatchNormalization())

model.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), padding='same', return_sequences=True))
model.add(BatchNormalization())

model.add(ConvLSTM2D(filters=40, kernel_size=(3, 3), padding='same', return_sequences=True))
model.add(BatchNormalization())

model.add(Conv3D(filters=1, kernel_size=(3, 3, 3), activation='relu', padding='same', data_format='channels_last'))

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Print out the model summary
model.summary()
```

```python
# Fit and save the model 
history = model.fit(X_train, y_train, batch_size=2, epochs=30, validation_split=0.3)
model.save('/content/drive/My Drive/Ubiik/model.h5')

# Create a dataframe to collect the history and save it
df_history = pd.DataFrame(history.history)
df_history.to_csv('/content/drive/My Drive/Ubiik/history.csv')
df_history.head()
```

```python
def plot_loss(df_history):

    """This function is to plot the loss curves for training and validation data

    Args:
        history: history dataframe of the model training
        n_epoch: number of epoch in the history

    """
    
    n_epoch = len(df_history)
    plt.figure(figsize=(10,6))
    plt.plot(np.linspace(1,n_epoch, n_epoch), df_history['loss'], '-o', label='Train')
    plt.plot(np.linspace(1,n_epoch, n_epoch), df_history['val_loss'], '-o', label='Validation')
    plt.xlabel('Epoch', size=12)
    plt.ylabel('MSE', size=12)
    plt.legend()
    plt.title('Loss Curve Tracking', size=18);
```
